```python
import pandas as pd
datos = pd.read_csv("SOCR-HeightWeight.csv")
datos = datos[['Talla','Peso']]
datos['Peso']= datos['Peso']*0.45359237
datos['Talla']= datos['Talla']*2.54
```

```python
datos.head()
```

```python
import seaborn as sb
sb.scatterplot(x="Talla", y="Peso", data= datos,
                hue="Peso", palette="coolwarm")
```

```python
X=datos['Talla']
Y=datos['Peso']
X_procesada=X.values.reshape(-1,1)
Y_procesada=Y.values.reshape(-1,1)
```

```python
Y_procesada
```

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
```

```python
modelo.fit(X_procesada, Y_procesada)
```

```python
talla = 166
prediccion = modelo.predict([[talla]])
print(f'{talla} centimetros nos da un prediccion de {prediccion} kilos')
```

```python
modelo.score(X_procesada, Y_procesada)
```
